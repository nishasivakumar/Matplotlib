```python
x = ["maths","english","tamil","science","civis","history"]
y = [300,500,450,200,350,250]

#you can give different colour for each bars too
c = ["red","green","blue","black","magenta","yellow"]
plt.bar(x,y,width=0.5,color =c,linewidth= 20)

#plt.bar(x,y,width=0.5,color ="red")
plt.xlabel("subjects")
plt.ylabel("no. of students passed")
plt.title("student details")
plt.show()
```

```python
x = ["maths","english","tamil","science","civis","history"]
y = [300,500,450,200,350,250]

plt.bar(x,y,width=0.5,color="green",linewidth=20,edgecolor="red")  #arguments --> width/color/linewidth  (etc)



plt.xlabel("subjects")
plt.ylabel("no. of students passed")
plt.title("student details")
plt.show()
```

```python
#Horizontal Bar graph

#we need to us "barh" function

x = ["maths","english","tamil"]
y = [300,500,450]
#you can give different colour for each bars too
c = ["green","magenta","yellow"]
plt.barh(x,y,[0.1,0.5,0.3],color =c,linewidth= 20) #[0.1,0.5,0.3] are width
plt.xlabel("subjects")
plt.ylabel("no. of students passed")
plt.title("student details")
plt.show()
```

```python
#stacked bar chart 

#function --> bar
#bottom 

#you can use this for categorical values like gender or any
#"bottom" parameter is very important

x = ["usa","india","china","russia","germany"]
gold =[3,7,5,4,2]
silver=[30,20,45,12,34]
bronz=[50,60,55,71,45]

plt.bar(x,gold,0.4,label="Gold")
plt.bar(x,silver,0.4,bottom=gold,label="Silver")
plt.bar(x,bronz,0.4,bottom=silver,label="Bronz")

plt.xlabel("Countries")
plt.ylabel("medals")
plt.title("no.of medals each country received")
plt.legend()
plt.show()


    
```

```python
#stacked graph

x =["CSE","IT","EEE","Mech"]
boys =[20,10,30,50]
girls=[60,45,12,2]

plt.bar(x,boys,0.4,label="Boys")
plt.bar(x,girls,0.4,bottom=boys,label="Girls")

plt.xlabel("Department")
plt.ylabel("strength of the students")
plt.title("no.of girls and boys in each department")
plt.legend()
plt.show()


```

```python
#Multiple Bar chart / Grouped Bar chart

#its similar to stacked bar graph but "it will place next to another"

#whereas stacked bar "it will place one above the another"

#stacked bar - it will place one above the another
#Multiple bar - it will place next to another
```

```python
import numpy as np

x =["CSE","IT","EEE","Mech"]
boys =[20,10,30,50]
girls=[60,45,12,2]
w =0.4
bar1 = np.arange(len(x))
bar2 = [for i in bar1]

plt.bar(bar1,boys,w,label="Boys")
plt.bar(bar2,girls,w,label="Girls")

plt.xlabel("Department")
plt.ylabel("strength of the students")
plt.title("no.of girls and boys in each department")
plt.legend()
plt.show()
```

```python
#Scatter Plots

#The scatter plot displays values for, typically, two variables for a set of data. 
#Such kind of plots could be very informative when figuring out relationships between pairs of variables.


import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("C:\\Users\\Athiban\\Desktop\\Kee\\Salary_data.csv")  # load dataset

X = data["YearsExperience"]
Y = data["Salary"]

plt.scatter(X, Y)
plt.title("Scatter Plot")
plt.xlabel("Working Experience (years)")
plt.ylabel("Annual Wage (dollars)")
plt.show()
```

```python
#We are also able to, for example, 
#distinguish between observations that have more than 5 years of working experience
#and observations that have less than 5 years of working experience by using different colors.
#To do this, we create two scatter plots by using the relevant data splits and display them in one single plot. 
#The following code results in the desired plot:
```

```python
X_1 = X[X > 5]
X_2 = X[X <= 5]
Y_1 = Y[X > 5]
Y_2 = Y[X <= 5]

plt.scatter(X_1, Y_1, label='Years of experience > 5')
plt.scatter(X_2, Y_2, label='Years of experience <= 5')
plt.title("Scatter Plot (split)")
plt.legend()
plt.xlabel("Working Experience (years)")
plt.ylabel("Annual Wage (dollars)")
plt.show()
```

```python
#Histogram Example 

#To give an example, we make use of the Iris dataset which is available through the sklearn package in Python. 
#Within this dataset, observations could belong to three different flower (iris) classes. 
#In this example, we consider only two features so that the picture is 2D. 
#These two features of interest are the sepal length and sepal width (in cm).

#what is iris dataset?
#Iris dataset contains five columns such as Petal Length, Petal Width, Sepal Length, Sepal Width and Species Type. 
#Iris is a flowering plant, 
#the researchers have measured various features of the different iris flowers and recorded digitally.

from sklearn import datasets
import matplotlib.pyplot as plt

bins = 20
iris = datasets.load_iris()
X_iris = iris.data

fig, axs = plt.subplots(2, 2)
axs[0, 0].hist(X_iris[:, 0])
axs[0, 1].hist(X_iris[:, 1], color='orange')
axs[1, 0].hist(X_iris[:, 2], color='green')
axs[1, 1].hist(X_iris[:, 3], color='red')

i = 0
for ax in axs.flat:
    ax.set(xlabel=iris.feature_names[i], ylabel='Frequency')
    i += 1

fig.suptitle("Iris Histograms")
```

```python
from sklearn import datasets
import matplotlib.pyplot as plt
bins = 20
iris = datasets.load_iris()
X_iris = iris.data
X_sepal = X_iris[:, 0]

plt.hist(X_sepal, bins)
plt.title("Histogram Sepal Length")
plt.xlabel(iris.feature_names[0])
plt.ylabel("Frequency")
plt.show
```
