```python
import matplotlib.pyplot as plt
x = [1, 2, 3]
x2 = [1, 3, 3.5]
y = [-4, 0, 8]
y2 = [-2, 3, -1]

plt.plot(x, y, label='Line 1')
plt.plot(x2, y2, label='Line 2')

plt.title("Plotting two simple lines")
plt.grid(True)
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.xlim([0, 4])
plt.ylim([-5, 10])
plt.legend()
plt.show()
```

```python
#Matplotlib object oriented method

#2 instance --> creating a figure instance and can add axes to that figure

 
#create a empty figure
#add axes to that figure

#create a empty figure
fig = plt.figure()

#do add_axes to that figure
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left,bottom,width,height  -->here i gave default numbers, but you can give whatever you want
#axes = fig.add_axes([-0.5,-0.5,0.6,0.6]) 

axes.plot(x,y,'b')
axes.set_xlabel("set x-axis")  #we need to use set_ --> for object oriented method
axes.set_ylabel("set y-axis")
axes.set_title("set title")

plt.show()
```

```python
#create a empty canvas/figure

fig = plt.figure()

#add_axes to that figure
#and we can easily add more than one axis to the figure

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #main axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #inset axes

#main axes | larger axes

axes1.plot(x,y,'g')
axes1.set_xlabel('x1_axis')
axes1.set_ylabel('y1_axis')
axes1.set_title('main plot | larger plot')

#inset axes | smaller axes

axes2.plot(y,x,'r')
axes2.set_xlabel('x2_axis')
axes2.set_ylabel('y2_axis')
axes2.set_title('inset plot | smaller plot')
```

```python
#subplots 

#this plt.subplots() act as a more automatic  axis manager 

#use similar to plt.figure() except use tuple unpacking to grab fig and axes

#what is Tuple unpacking?

#in python there is a dictionary called tuple eg: t = (1,2,3)

#    tuple           -->    t = (1,2,3)
#    tuple packing   -->    t =  1,2,3
#    tuple unpacking -->   a,b,c = (1,2,3)


fig, axes = plt.subplots()

#now use axes object and stuff to plot

axes.plot(x,y,'b')
axes.set_xlabel('X_Axis')
axes.set_ylabel('Y_Axis')
axes.set_title('Ex of subplots')
```

```python
#then you can specify the rows and columns when creating the subplots object

#Empty canvas of 1 by 2 subplots

#Note : subplots will create the axes automatically in background(thats why we are creating the figure and axes, simply we are giving the name)

fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
#here the common issues is overlapping subplots or figure
#in this case, we can use fig.tight.layout() or plt.tight.layout() method, wh automatically adjust position axes on the figure canvas
#                          ----------------------------------------
#so there is no overlapping content

#tight_layout()  --> automatically adjust subplot parameters 

fig, axes = plt.subplots(nrows=1, ncols = 2)

for i in axes:
    i.plot(x,y,'r--')
    i.set_xlabel('X_axis')
    i.set_ylabel('Y_axis')
    i.set_title('title')
fig
plt.tight_layout() 

```

```python
#Figure size, aspect ratio and Dpi

#Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created, 
#using the figsize and dpi keyword arguments. figsize is a tuple of the width and height of the figure in inches, 
#and dpi is the dots-per-inch (pixel per inch). 
#To create an 800x400 pixel, 100 dots-per-inch figure

fig = plt.figure(figsize=(2,1), dpi=100)
```

```python
#The same arguments can also be passed to layout managers, such as the subplots function

fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x,y,'r')
axes.set_xlabel('X_axis')
axes.set_ylabel('Y_axis')
axes.set_title('title')
```

```python
#Formatting text: LaTeX, fontsize, font family

#there is format called "LaTeX"  --> you need to give $, please see the example below 

#eg:LaTeX in any text (legend, title, label, etc.). For example, "$y=x^3$"

#In LaTeX, we frequently use the backslash \ in commands
#for example \alpha to produce the symbol  α . 
#Raw text strings are prepended with an 'r',
#like r"\alpha" or r'\alpha' instead of "\alpha" or '\alpha'

#you need to give "$" in start and end, (start, need to give after "r")


fig, axes = plt.subplots()

axes.plot(x,x**2, label=r'$y = \alpha^2$') #y = a2  
axes.plot(x,x**3, label=r'$y = \alpha^3$') #y = a3
axes.legend(loc=1)
axes.set_xlabel(r'$\alpha$', fontsize=18)
axes.set_ylabel(r'$y$',fontsize=18)


          


```
